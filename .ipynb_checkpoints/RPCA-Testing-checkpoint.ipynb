# Testing RPCA-PGSG on tree dataset
Reads in ./data/trees.bin

Writes output to ./results/

```julia
include("./PGSG.jl")
using PGSG
using PyPlot
```

```julia
##### load data #####
println("load data...")
n,m = 20480, 102;
fid = open("./data/trees.bin","r")
A   = read(fid, UInt8, n*m)
A   = reshape(A, n, m)
close(fid)

A   = A/255

##### set up parameters #####
k   = 10; #number of componets in low rank model

# svd initialization
println("svd initialization...")
UΣV,nconv,niter,nmult,resid = svds(A,nsv=k,tol=1e-12)
U   = diagm(sqrt(UΣV[:S]))*UΣV[:U].';
V   = diagm(sqrt(UΣV[:S]))*UΣV[:V];
for I in eachindex(U)
    U[I] < 0.0 ? U[I] = 0.0 :
    U[I] > 1.0 ? U[I] = 1.0 : continue
end
for I in eachindex(V)
    V[I] < 0.0 ? V[I] = 0.0 :
    V[I] > 1.0 ? V[I] = 1.0 : continue
end
```

```julia
##### apply solver #####
println("start solver...")
T=64
sol,obj = RPCA_PGSG(U,V,A,0.25,2.0,T, 1.5); 
U,V = sol[1],sol[2]
```

```julia
min =123
new_obj = zeros(convert(Int64, (T-1)*(T-2)/2))
for i in 1:convert(Int64, (T-1)*(T-2)/2)
    new_obj[i] = log((obj[i]-min)/min)
end

plot(1:convert(Int64, (T-1)*(T-2)/2), new_obj)
xlabel("Inner Iterations")
ylabel("Objective")
savefig("temp-ObjectivePlot.png")
```

# Write output to Results directory

```julia
##### recover S and save data #####
m,n = size(A);
println("recover S...")
L = U.'*V;
r = A - L;
κσ = 0.35;
# κσ = κ;
S = zeros(m,n);
for I in eachindex(S)
    r[I] >  κσ ? S[I] = r[I] - κσ :
    r[I] < -κσ ? S[I] = r[I] + κσ : continue;
end

##### save results #####
println("save data...")
fid = open("./results/trees_PGSG_UV.bin","w");
write(fid,U,V);
close(fid);
fid = open("./results/trees_PGSG_LS.bin","w");
write(fid,L,S);
close(fid);
fid = open("./results/trees_PGSG_R.bin","w");
write(fid,r);
close(fid);
```

```julia
function oneNorm(A)
    r=0
    n,m = size(A)
    for i in 1:m
        for j in 1:m
            r = r + abs(A[i,j])
        end
    end
    return r
end
function test_RPCA_subgradient(U,V,A,T, scale=1)
    function stochastic_subgradient1(X)
        U,V = X[1],X[2]
#        R = U'*V
        
        n,m = size(U)
        n,k = size(V)
        ∇U = zeros(n,m)
        ∇V = zeros(n,k)
        
        i = rand(1:m) #for i in 1:m #iterate over pixels
            j = rand(1:k) #random over frames
                if (U[:,i]'V[:,j])[1] - A[i,j]>0.001 #R[i,j]
                    ∇U[:,i] = ∇U[:,i] + V[:,j]
                    ∇V[:,j] = ∇V[:,j] + U[:,i]
#                    for l in 1:n #iterate over components
#                        ∇U[l,i] = ∇U[l,i] + V[l,j]
#                        ∇V[l,j] = ∇V[l,j] + U[l,i]
#                    end
                end
                if (U[:,i]'V[:,j])[1]  - A[i,j]<-0.001
                    ∇U[:,i] = ∇U[:,i] - V[:,j]
                    ∇V[:,j] = ∇V[:,j] - U[:,i]
#                    for l in 1:n #iterate over components
#                        ∇U[l,i] = ∇U[l,i] - V[l,j]
#                        ∇V[l,j] = ∇V[l,j] - U[l,i]
#                    end
                end
        #end
        return [∇U, ∇V] *scale
    end

    obj = zeros(T)
    alpha=0.013
    X = [U,V]
    for j in 1:T
        X = X - (alpha/(j)^(0.2))*stochastic_subgradient1(X)
        obj[j] = 100#oneNorm(X[1].'*X[2] - A)
        if j%2500==0 println("Inner loop finished ", oneNorm(X[1].'*X[2] - A)) end
    end
    return X, obj
end
```
